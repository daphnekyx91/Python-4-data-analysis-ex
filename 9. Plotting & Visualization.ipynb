### Adjusting the spacing around subplots
- wspace, control the percent of figure width
- hspace, control the percent of horizontal width
- it control the spacing between the subplots

```python
figs, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.3 )
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
#change the % of wspace and hspace

figs, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.3 )
plt.subplots_adjust(wspace=0.2, hspace=0.2)
```

### Ticks, Labels and Legends

### Setting the title, axis labels, ticks and ticklabels

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
plt.title('Simple plot illustrating xticks')
```

```python
# insert xticks
ticks = ax.set_xticks([0, 250, 500, 750, 1000])

# insert labels
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                           rotation=30, fontsize='small')

#insert title
ax.set_title('My first matplotlib plot')

#insert xlabel
ax.set_xlabel('Stages')
```

```python
from datetime import datetime

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='g-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankrupcy')
]

for date, label in crisis_data:
    ax.annotate(label , xy=(date, spx.asof(date) + 75),
                xytext = (date, spx.asof(date) + 225),
                arrowprops = dict(facecolor='red', headwidth=4, width= 2,
                                 headlength= 4),
                horizontalalignment='left', verticalalignment='top')
    
    
#Zoom in on 2007 - 2010
# set the xlim and ylim

ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([ 600, 1800 ])

ax.set_title('Important dates in the 2008 - 2009 financial crisis')
```

```python
# add patch object/ shapes using object shp

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

### Bar Plots

- plot.bar() make a vertical bar plot
- plot.barh() make a horizontal bar plot
Series or DataFrame index will be used as the x(bar) or y(barh) ticks

```python
fig, axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))


data.plot.bar(ax =axes[0], color='g', alpha=0.7)
data.plot.barh(ax=axes[1], color='pink', alpha=0.7)
```

### Histograms and Density Plots

- bar plot to give a descrtixed display of value frequency
- data points are split into discrete, evenly spaced bins and number of data points in each bin is plotted
- use plot.hist method


```python
tips['tip_pct'].plot.hist(bins=50)
```

```python
#plot density plot
# also known as kernel density estimate plots

tips['tip_pct'].plot.density()
```

```python
# distplot can plot both histogram and a continuous density estimate
# simultaneously

comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)

values = pd.Series(np.concatenate([comp1, comp2]))

sns.distplot(values, bins=100, color='g')
```

### Scatter or Point plots

- Point plots or scatter plot is a useful way of examining relationship
- between 2 1-dimensional data series

- load the macrofata from the statsmodel project

```python
macro = pd.read_csv('macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]

trans_data = np.log(data).diff().dropna()

trans_data[-5:]
```

```python
# use seaborn's regplot method to make scatter plot
# and fit the linear regression line

sns.regplot('m1', 'unemp', data=trans_data)

plt.title('Changes in log %s versus log &s' % ('m1', 'unemp'))
```

```python
# Exploratory data analysis to look at the scatter plot
# known as pair plots or scatter plot matric
# Seaborn has a convenient pairplot function


sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

### Facet Grids & Categorical Data

- have additional grouping dimensions
- visualize data with many categorical variables - use facet grid
- factorplot simplifies using faceted plots



```python
sns.factorplot(x='day', y ='tip_pct', hue='time', col='smoker',
              kind='bar', data=tips[tips.tip_pct < 1])
```

```python
# grouping by 'time' by different bar colors within a facet
# expand the facet grid by adding one row per time value

sns.factorplot(x='day', y='tip_pct', row='time',
              col= 'smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
# create box plots

sns.factorplot(x='tip_pct', y ='day', kind='box',
               data = tips[tips.tip_pct < 0.5])
```
